### 10. Exceptional Handling

An `exception` is an unexpected `event/error` that occurs during the execution of a program. 

```python
a = 10
b = 0

print(a/b)  # it throuw an error and your program terminate
print('Hello world')
```

### 2. Types of Errors in Python

Pyton errors are mainly of two types:


A.Syntax Error(Compile time error)

Occurs when python cann't understand your code

B.Runtime Error(Exception)

Occurs while executing the program

### Why Exception Handling is Required

### Without Exception Handling:
* Program crashes
* user gets error message while bug in your code
* system become unreliable.
### With Exception Handling:
* Program continues after error
* We can show user-friendly message
* better debugging and logging
* prevents data loss

### 4. Exception Handling Keywords

Python provides these keyword

|Keyword|Meaning|
|---|---|
|try|codes that may generate errors/event|
|except|handles exception|
|else|runs if no exception|
|finally|always runs|
|raise|manually generate exceptions|
|assert|testing conditions, raise AssertionError|

## Basic Syntax of try-except

```python
try:
    a=10
    b = 0
    print(a/b) # error
except:
    print("Cannot divide by zero")
print("Hello world")
```

###  Handling specific exceptions

Always prefer handling specific exception,not general.

```python
try: 
    a= 10
    b=0
    print(a/b)
except ZeroDivisionError:
    print("Division by zero is not allowed")
except TypeError:
    print("Check datatype of the value")
```
