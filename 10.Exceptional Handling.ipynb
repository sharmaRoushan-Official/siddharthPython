### Why Exception Handling is Required

### Without Exception Handling:
* Program crashes
* user gets error message while bug in your code
* system become unreliable.
### With Exception Handling:
* Program continues after error
* We can show user-friendly message
* better debugging and logging
* prevents data loss

Multiple Except Blocks

You can handle multple errors at a time.

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)
except ValueError:
    print("Invalid input! Please enter only a number")
except ZeroDivisionError:
    print("You cannot divide by zero")
```

#### single Except Handling Multiple Exceptions

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)
except(ValueError,ZeroDivisionError):
    print("Invalid input or Division by Zero")
```

### Using Exceptions as Object(as e)


```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)
except Exception as e:
    print("Error occurred:",e)
```

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(x/y)
except ZeroDivisionError:
    print("ZeroDivisionError")
except Exception as e:
    print(e)
```

#### try-block-else Block

else block runs only if not exception occurs.

```python
try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    z = x/y
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division Result:",z)
```

* if exception -> except runs
* if no excption -> else runs

try-except-else-finally Block

finally block alwasy runs(wheter exception happens or not)

```python
try:
    with open(r"‪C:\Users\User\Desktop\NursaryRhymes.txt",'r') as f:
        content = f.read()

except Exception as e:
    print(e)
else:
    print(content)
finally:
    print("Program finished execution")
```

### Nested Try:Except

```python
try:
    a = int(input("Enter A:"))
    b = int(input("Enter B:"))
    try:
        print(a/b)
    except ZeroDivisionError:
        print("Inner Block:division by zero")
except ValueError:
    print("Outer Block:invalid input")
```

### Using raise Keyword (manually throw Exception)

Sometimes you want to create your own error



```python
age = int(input("Enter age:"))
if age<=18:
    raise ValueError("Age must be 18 or above")
print("Elegible")
```

Custom Exceptions

You can create your own Exception

```python
class InvalidAgeError(Exception):
    pass
```

```python
age = int(input("Enter age:"))
if age<=18:
    raise InvalidAgeError("Age must be 18 or above")
print("Elegible")
```

```python
try:
    age = int(input("Enter age:"))
    if age<=18:
        raise ValueError("Age must be 18 or above")
except Exception as e:
    print(e)
else:
    print("Elegible")
```

```python
class InvalidAgeError3(Exception):
    pass
```

```python
try:
    age = int(input("Enter age:"))
    if age<=18:
        raise InvalidAgeError3("Age must be 18 or above")
except Exception as e:
    print(e)
else:
    print("Elegible")
```
