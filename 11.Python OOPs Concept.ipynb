## 7. Types of Methods 
1. Instance Method 
    > Used Self
    > Access Instance variables
2. Class Methods
   > Uses @classmethod
   > uses cls
3. Static Method
   > Uses @staticmethod
   > No self of cls

```python
# Wap to take length and breadth of rectangle and display their area and perimeter.

class Rectangle:
    name = "apple" # class variable
    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth

    def detailsRectangle(self):
        area = self.length*self.breadth
        print(f"Area of Rectangle is:{area} Cm²")
        perimeter = 2*(self.length+self.breadth)
        print(f"Perimeter of Rectangle is:{perimeter}")

rect1 = Rectangle(10,20)
rect1.detailsRectangle()
```

### 1. Instance Method
* Used Self Access Instance variables

* An instance method is the most common method in a class.

* Uses self as the first parameter

* works with object-specific data

* Can access and modify instance variables

```python
class Student:
    def __init__(self,name,marks):
        self.name = name # instance variables
        self.marks = marks # instance variable
    def display_details(self):
        print("Student Name:",self.name)
        print("Student Marks:",self.marks)

s1 = Student("Ravi",86)
s1.display_details()
```

## 2. Class Mehtod

A `class method` works with the class-level data, not object-level

### Features:
* Uses @classmethod decorator
* Uses cls as the first parameter
* Used to access/modify class variables
* Can be called using class name or object 

```python
class Company:
    company_name = "Google" # class varibale
    def __init__(self,employee):
        self.employee = employee # instance variable

    @classmethod
    def change_company_name(cls,new_name):
        cls.company_name = new_name

print("Before:",Company.company_name)
Company.change_company_name("Yahoo")
print("After:",Company.company_name)
```

3. Static Method 

A `static method` is a mthod that does not use self or cls

### Features:
* Uses @staticmethod decorator
* No self or cls
* Used for utility/helper functions
* Cannot access instance variable or class variables directly.

```python
class Calculator:

    @staticmethod   
    def add(a,b):
        return a+b

result = Calculator.add(10,20)
print("Addition=",result)
```

add() mehtod is a static method beacuase it does not uses cls or self.

It behave like a normal function but is kept inside the class.

it is used for general-purpose operation related to the class.
