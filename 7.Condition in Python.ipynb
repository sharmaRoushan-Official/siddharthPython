A conditonal statments allows a program to make decision based on conditions(True/False)

```python
age = 6

if age>5:
    print("You can go to skool")  # True block
```

```python
age = 5

if age>5:
    print("You can go to skool")  # True block
else:
    print("You are a baby") # False block
```

```python
# if - elif - else statement 

age = 4

if age>5:
    print("You can go to primary skool")
elif age == 5:
    print("You can go to kindergarten")
else:
    print("you are baby")
```

```python

age = -10

if age>5:
    print("You can go to primary skool")
elif age == 5:
    print("You can go to kindergarten")
elif age<5 and age>0:
    print("You are a baby")
else:
    print("Invalid age")
```

```python
# <!-- nested if  -->



age = 10

if age>0 and age<=100:
    if age>5:
        print("You can go to primary skool")
    elif age == 5:
        print("You can go to kindergarten")
    
    else:
        print("You are a baby")
else:
    print("Enter a valid age")
```

```python
## wap to take input from user if user press 1-"Monday"  7 = Saturday, 8 = "Sunday" , 10 - Invalid Weekday


week = int(input("Enter any number from 1-8: "))

if week>0:
    if week == 1:
        print("Monday")
    
    elif week == 2:
        print("Tuesday")
    elif week == 3:
        print("Wednesday")
    elif week == 4:
        print("Thursday")
    elif week == 5:
        print("Friday")
    elif week  == 6:
        print("Saturday")
    elif week == 7:
        print("Sunday")
    else:
        print("Invalid Weekday Number")
else:
    print("Weekday cannot in Negative value.")
```

```python
# wap to find largestst of three numbers

num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))

if num1>num2 and num1>num3:
    print(f"{num1} is greater than {num2} and {num3}")
elif num2>num1 and num2>num3:
    print(f"{num2} is greater than {num1} and {num3}")
elif num3>num1 and num3>num2:
    print(f"{num3} is greater than {num1} and {num2}")
else:
    print("Enter a valid number")
```
