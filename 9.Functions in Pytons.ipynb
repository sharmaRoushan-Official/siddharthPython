## 7. Scropes of Variables 

1. Local Variable

````python
def test():
    x = 10  # local variable 
    print(x)

test()
````

````python
#  Global Variable 

x = 100 # Global variable

def test():
    print(x)
test()
````

````python
## Nested Functions

def outer():  # outer functions
    def inner(): # inner functions
        print("Inside inner functions")
    inner()

outer()
````

````python

def outer():  # outer functions
    def inner(): # inner functions
        print("Inside inner functions")
        def inner2():
            print("This is inner 2 functions")
        inner2()
    inner()
outer()
````
