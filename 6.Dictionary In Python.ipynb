### Definition of Dictionary


A dictionary is written using curly brackes {} and contatins elemnets in format.

```python
data = {
    "name":"Ram Kumar",
    "age": 25,
    "email":"ramkumar@gmail.com"
}
```

```python
data
```

```python
data = {"name":"Ram Kumar","age": 25,"email":"ramkumar@gmail.com"}
data
```

'name' is key -> "Ram kumar" is the value

"age" is key -> 25 is the value.

"email" is key -> ramkumar@gmail.com is the the value.

### Key feature of dicitonary

1. key-Value pair sturucture

2.Key must be unique

3.Value can be duplicates

4.Dictionary is mutable

5. dictionary are unordered (3.7+) -

### Creating a dictionary

Method: 1 Using curly braces {}

```python
data1 = {"name":"Rohan singh","age":35,"address":"Punjab"}
print(data1,type(data1))
```

method: 2 

using Constructor dict() 

```python
data2 = dict(name="Rohan singh",age=35,address="Punjab")
print(data2,type(data2))
```

Method 3

To create empty dictionary

```python
d = {}
print(d,type(d))
```

## Nested Dictionary 

A dictionary inside other dicitonary is called nested dictionary.

```python
employee = {
    101:{
        "name":"Binod",
        "age":35,
        "address":"Noida"
    },
    102:{
        "name":"Durgesh",
        "age":36,
        "address":"Bhopal"
    }
}
print(employee)
```

```python
employee[101]
```

```python
employee
```

```python
employee[101]['address']
```

```python
employee[102]['age']
```

```python
employee = {
    101:{
        "name":"Binod",
        "age":35,
        "address":"Noida",
        'phone':{
            "ph1":123456789,
            "Ph2":789456123
        }
    },
    102:{
        "name":"Durgesh",
        "age":36,
        "address":"Bhopal"
    }
}
print(employee)
```

```python
employee[101]['phone']["Ph2"]
```

```python
employee['ph2']
```
